## Collecting All tickers used in S&P 500 after 1st Jan 2000

```python
stock_part_info = pd.read_csv("S&P 500 Composition Changes 1970 - 2024 - SP 500 Summary Sheet.csv")
stock_part_info.head()
```

Change column name to string represantation of date

```python
rename_cols = {
    cur_name : datetime.strptime(cur_name, "%m/%d/%Y").date()
    for cur_name in stock_part_info.columns
    if cur_name not in ["Ticker", "Company Name", "ISIN Code"]
}
stock_part_info.rename(columns=rename_cols, inplace=True)
stock_part_info.head()
```

Remove all unessasory columns

```python
keep_cols = [col for col in stock_part_info.columns if col in ["Ticker", "Company Name", "ISIN Code"] or col >= datetime(1999, 12, 1).date()]
stock_part_info = stock_part_info[keep_cols]
stock_part_info.head()
```

Remove all rows which are all X (They weren't part of S&P 500 post Jan 1st)

```python
stock_part_info = stock_part_info[stock_part_info.iloc[:, 3:].notnull().any(axis=1)]
stock_part_info.head()
```

```python
all_stock_info_needed = stock_part_info[["Ticker", "Company Name", "ISIN Code"]]
all_stock_info_needed.head()
```

Saving all stock ticker who's information has to be collected for the enviroment set up

```python
all_stock_info_needed.to_csv("../all_stock_info_needed.csv", index=False)
```

## Collecting All tickers used in S&P 500 on 1st Jan 2000

```python
sp_500_stock_on_beginning = stock_part_info[["Ticker", "Company Name", "ISIN Code", datetime(1999, 12, 31).date()]]
sp_500_stock_on_beginning.head()
```

Remove all stocks which weren't part of s&p 500 on Jan 1st 2000

```python
sp_500_stock_on_beginning = sp_500_stock_on_beginning[sp_500_stock_on_beginning[datetime(1999, 12, 31).date()].notnull()]
sp_500_stock_on_beginning.head()
```
